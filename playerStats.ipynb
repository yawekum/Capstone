```python
# Importing libraries
import pandas as pd
import csv

# Reading CSV file
players = pd.read_csv('player_stats.csv', header=0, encoding='unicode_escape')
#print(players.to_string())

# Check number of nulls in each column
players.isnull().sum()

# Fill null with empty spaces " " into a new dataframe called modifiedPlayers
modifiedPlayers = players.fillna(" ")
#print(modifiedPlayers.to_string())

# Display columns
modifiedPlayers.columns

# Rename colums
modifiedPlayers1 = modifiedPlayers.rename({'90s': 'Per90', 'Compition': 'League', 'Born year': 'Birth Year', 'Match played': 'Matches Played', 'MIn': 'Minutes Played', 'Goal + Assist':'Goal n Assist', 'Goal+Penalty kick': 'Goal n Penalty Kick', 'Progressive Carries':'Carries', 'Progressive Passes':'Passes', 'Progressive Passes Receive': 'Passes Received', 'Non-penalty Expected Goal': 'NonPenalty Expected Goal' }, axis=1)
#print(modifiedPlayers1.to_string())

# Remove columns
modifiedPlayers2 = modifiedPlayers1.drop(['Penalty Kick','Penalty Kick Attempted', 'Red Card', 'Yellow Card','Non-penalty Expected Goal + Expected Assisted Goal'], axis=1)
print(modifiedPlayers2.to_string())

# Write to CSV without index column
modifiedPlayers2.to_csv('CleanedPlayerStatsPython.csv', index=False)
```
